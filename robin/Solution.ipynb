```python
class MeanEncoder:
    
    def __init__(self, df, on, by, prior_count=100):
        self.on = on
        self.by = by
        counts = df.groupby(by)[on].agg(['mean', 'count'])
        avg = df[on].mean()
        self.means = (
            counts
            .eval('(mean * count + @avg * @prior_count) / (count + @prior_count)')
            .rename(str(self))
        )
    
    def __str__(self):
        return f'avg_{self.on}_by_{self.by}'
        
    def transform(self, df):
        return df.join(self.means, on=self.by)[self.means.name]
```

```python
class CountEncoder:
    
    def __init__(self, df, on):
        self.on = on
        self.counts = df.groupby(on).size().rename(str(self))
    
    def __str__(self):
        return f'{self.on}_size'
        
    def transform(self, df):
        return df.join(self.counts, on=self.on)[self.counts.name]
```

```python
class NMissing:
    
    def __str__(self):
        return 'n_missing'
    
    def transform(self, df):
        return df.isnull().sum(axis='columns').rename(str(self))
```

```python
extractors = [
    MeanEncoder(df, on='target', by='addr_region_fact'),
    CountEncoder(df, 'app_addr_region_sale_encoding1'),
    CountEncoder(df, 'app_addr_region_sale_encoding2'),
    CountEncoder(df, 'inquiry_recent_period'),
    NMissing(),
    MeanEncoder(df, on='missing', by='addr_region_fact'),
]
```

```python
import catboost
from scipy import special

fit = catboost.Pool(X_fit, y_fit, cat_features=cat_cols)
val = catboost.Pool(X_val, y_val, cat_features=cat_cols)

model = catboost.CatBoostClassifier(iterations=500,
                                    learning_rate=0.01,
                                    depth=8,
                                    eval_metric='AUC',
                                    random_seed = 42,
                                    bagging_temperature = 0.2,
                                    od_type='Iter',
                                    metric_period = 50,
                                    od_wait=50)

model.fit(fit, 
          use_best_model=True,
          eval_set=val)
```

```python
import numpy as np
from sklearn import metrics

y_pred = model.predict_proba(X_val)[:, 1]

def roc_auc_score_at_K(predicted_proba, target, rate=0.1, quantile_inf=0.05, quantile_sup=0.95):      
    predicted_proba = pd.Series(predicted_proba)
    quantile_inf = predicted_proba.quantile(quantile_inf)
    quantile_sup = predicted_proba.quantile(quantile_sup)
    predicted_proba = predicted_proba.apply(lambda x: 10**-6 if x < quantile_sup and x > quantile_inf else x)
    order = np.argsort(-predicted_proba) 
    top_k = int(rate * len(predicted_proba)) 
    return metrics.roc_auc_score(target[order][:top_k], predicted_proba[order][:top_k]) 

val_score = roc_auc_score_at_K(y_pred, y_val)
val_score
```
